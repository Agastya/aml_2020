```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
import matplotlib.pyplot as plt
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename == 'fire_nrt_V1_96617.csv':
            file_path = os.path.join(dirname, filename)

# Any results you write to the current directory are saved as output.
dataset_raw = pd.read_csv(file_path)
print(dataset_raw.head())
# dataset_raw = dataset_raw.head(500)
```

```python

fig, axes = plt.subplots(2, 2, figsize=(18, 9))
axes[0,0].plot(dataset_raw['latitude'], dataset_raw['longitude'], 'o')
axes[0,1].scatter(dataset_raw['latitude'], dataset_raw['longitude'], alpha=0.01, lw=3)
axes[1,0].scatter(dataset_raw['latitude'], dataset_raw['longitude'],s=10)
hex_graph = axes[1,1].hexbin(dataset_raw['latitude'], dataset_raw['longitude'], bins='log')
fig.colorbar(hex_graph)

```

```python
saturated_data= dataset_raw.loc[dataset_raw['bright_ti4']>330]
unsaturated_data= dataset_raw.loc[dataset_raw['bright_ti4']<=330]
fig2, axes2 = plt.subplots(1, 2, figsize=(18, 9))
axes2[0].scatter(saturated_data['latitude'], saturated_data['longitude'], c=saturated_data['bright_ti4'], cmap='viridis')
axes2[1].scatter(unsaturated_data['latitude'], unsaturated_data['longitude'], c=unsaturated_data['bright_ti4'], cmap='viridis')
axes2[0].legend()
axes2[1].legend()
```
